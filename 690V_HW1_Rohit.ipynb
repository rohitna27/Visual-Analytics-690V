```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
```

```python
output_notebook()
```

```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile

df1 = pd.read_excel('C:/Users/Rohit/Desktop/690V/Indicator_Average age.xlsx', sheetname='Data')
```

```python
from bokeh.models import ColumnDataSource
source = ColumnDataSource(df1)
```

```python
from bokeh.models import HoverTool, BoxSelectTool
tools_used = [BoxSelectTool(), HoverTool()]
p = figure(plot_width=400, plot_height=400, tools=tools_used, title="Average ages of billionaires across 2004 and 2005")
p.circle('"2004"', '"2005"', source=source, size=5, color='navy', alpha=0.5)
```

```python
p.xaxis.axis_label = "Average ages in 2004"
p.yaxis.axis_label = "Average ages in 2005"
```

```python
show(p)
```

```python
from bokeh.charts import Line
from bokeh.models import HoverTool, Legend
from bokeh.palettes import Spectral6
import numpy as np
from itertools import cycle
```

```python
p = figure(plot_width=800, plot_height=400)
colors = Spectral6
colorcycler = cycle(colors)
for column in df1.columns:
    line = p.line(
        x=df1.index, y=column, source=source, legend="%s" % column,
        line_color=next(colorcycler), line_width=2
    )
    p.add_tools(HoverTool(tooltips=[("x,y","$x,$y"),("column", "&nbsp;%s" % column)], renderers=[line]))
show(p)
```

```python
from bokeh.charts import HeatMap
p = HeatMap(df1, title='Heat map of average ages of billionaires across countries')
# As a large number of points on the Heat map are at zero, we can infer that most conutries do not have any billionaires at all
# A heat map with fewer countries and greater clarity follows
show(p)
```
